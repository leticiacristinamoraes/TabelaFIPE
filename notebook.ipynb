```python
#streamlit run minha_aplicacao.py --server.headless true
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By   # para poder pegar elementos da pagina web
from selenium.webdriver.common.keys import Keys    #   para poder digitar botões do teclado paginas da web 
from selenium.webdriver.support.select import Select  # para poder selecionar opções de um menu drop down
from selenium.webdriver import ActionChains   # para mover o mouse para algum lugar e fazer alguma coisa (como abrir um menu drop down por exemplo)
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument("--start-maximized")
servico = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=servico,options=chrome_options)
```

```python
driver.get("about:blank")
```

```python
# Teste 1
time .sleep(3)
tab_metrics = driver.find_element(By.ID, 'tabs-bui2-tab-4')
tab_metrics.click()
time.sleep(6)
tab_select = driver.find_element(By.CLASS_NAME, 'st-key-nome_pesquisador')
list_select = tab_select.find_element(By.CSS_SELECTOR, 'div[data-baseweb="select"]')
selected = list_select.find_element(By.TAG_NAME, 'input')
selected.send_keys("Bob")
time.sleep(6)
selected.send_keys(Keys.ENTER)
time.sleep(10)

date_field_inicial = driver.find_element(By.CLASS_NAME, 'st-key-data_inicial')
date_inicial = date_field_inicial.find_element(By.TAG_NAME, 'input')
date_inicial.send_keys(Keys.CONTROL + "a")
date_inicial.send_keys("20241201")
date_inicial.send_keys(Keys.ENTER)
time.sleep(10)

date_field_final = driver.find_element(By.CLASS_NAME, 'st-key-data_final')
date_final = date_field_final.find_element(By.TAG_NAME, 'input')
date_final.send_keys(Keys.CONTROL + "a")
date_final.send_keys("20241231")
date_final.send_keys(Keys.ENTER)
time.sleep(15)

button_cotacoes = driver.find_element(By.CLASS_NAME, "st-key-key_calcular_comissoes")
button_cotacoes.find_element(By.TAG_NAME, 'button').click()
time.sleep(6)
```
